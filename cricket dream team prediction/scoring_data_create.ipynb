## Preparing the Batting data

```python
columns_and_types_batting = {
    "match_id": int,
    "date": str,
    "batter": str,
    "batter_runs": int,
    "balls_faced": int,
    "out": int
}

df_bat = pd.DataFrame(columns=columns_and_types_batting.keys()).astype(
    columns_and_types_batting
)
df_bat.info()
```

```python
def create_batting_sheet(df, df_bat):
    for index, row in df.iterrows():
        match_id = row["Match ID"]
        batter = row["Batter"]

        condition = (df_bat["match_id"].values == match_id) & (
            df_bat["batter"].values == batter
        )

        if row["Wicket"] == 0:
            new_row = {
                "match_id": match_id,
                "date": row["Date"],
                "batter": batter,
                "batter_runs": row["Total Batter Runs"],
                "balls_faced": row["Batter Balls Faced"],
                "out": row["Wicket"],
            }
            if condition.any():
                df_bat.loc[condition, new_row.keys()] = new_row.values()
            else:
                df_bat = pd.concat([df_bat, pd.DataFrame([new_row])], ignore_index=True)
        else:
            new_row = {
                "match_id": match_id,
                "date": row["Date"],
                "batter": batter,
                "batter_runs": row["Player Out Runs"],
                "balls_faced": row["Player Out Balls Faced"],
                "out": row["Wicket"],
            }

            df_bat.loc[condition, new_row.keys()] = new_row.values()

    return df_bat


df_bat = create_batting_sheet(df, df_bat)
df_bat.head()
```

```python
temp = (df_bat["batter_runs"] / df_bat["balls_faced"]) * 100
df_bat["strike_rate"] = temp

df_bat["date"] = pd.to_datetime(df_bat["date"])
df_bat_sorted = df_bat.sort_values(by="date")

df_bat_sorted.head()
```

```python
df_bat_sorted.to_csv("batting_custom.csv", index=False)
```

## Preparing the Bowling data

```python
columns_and_types_bowling = {
    "match_id": int,
    "date": str,
    "bowler": str,
    "wickets": int,
    "balls_bowled": int,
    "runs_conceded": int,
}
df_bowl = pd.DataFrame(columns=columns_and_types_bowling.keys()).astype(
    columns_and_types_bowling
)
df_bowl.info()
```

```python
def create_bowling_sheet(df, df_bowl):
    for index, row in df.iterrows():
        match_id = row["Match ID"]
        bowler = row["Bowler"]

        condition = (df_bowl["match_id"].values == match_id) & (
            df_bowl["bowler"].values == bowler
        )

        temp = df_bowl.loc[condition]

        if condition.any():
            wic = row["Wicket"] + temp["wickets"]
            runs = row["Bowler Runs Conceded"] + temp["runs_conceded"]
            balls = row["Valid Ball"] + temp["balls_bowled"]

            new_row = {
                "match_id": match_id,
                "date": row["Date"],
                "bowler": bowler,
                "wickets": wic,
                "balls_bowled": balls,
                "runs_conceded": runs,
            }

            df_bowl.loc[condition, new_row.keys()] = new_row.values()

        else:
            new_row = {
                "match_id": match_id,
                "date": row["Date"],
                "bowler": bowler,
                "wickets": row["Wicket"],
                "balls_bowled": row["Valid Ball"],
                "runs_conceded": row["Bowler Runs Conceded"],
            }
            df_bowl = pd.concat([df_bowl, pd.DataFrame([new_row])], ignore_index=True)
            df_bowl.head()

    return df_bowl


df_bowl = create_bowling_sheet(df, df_bowl)
df_bowl
```

```python
df_bowl.to_csv("bowling_custom.csv", index=False)
```

```python
bowling_columns = {
    "match_id": int,
    "date": str,
    "bowler": str,
    "wickets": int,
    "wicket_score": int,
    "balls_bowled": int,
    "runs_conceded": int,
    "maiden": int,
}
df_bowl = pd.DataFrame(columns=bowling_columns.keys()).astype(bowling_columns)


def create_bowling_sheet(df, df_bowl):
    for index, row in df.iterrows():
        match_id = row["Match ID"]
        bowler = row["Bowler"]

        condition = (df_bowl["match_id"].values == match_id) & (
            df_bowl["bowler"].values == bowler
        )

        temp = df_bowl.loc[condition]

        if row["Ball"] == 1:
            flag = True
            maiden = 0
        if row["Runs From Ball"] != 0:
            flag = False
        if row["Ball"] == 6 and flag == True:
            maiden = 1

        if condition.any():
            score = temp["wicket_score"]
            if pd.notna(row["Method"]):
                if row["Method"] != "run out":
                    score += 25
                if (row["Method"] == "lbw") | (row["Method"] == "bowled"):
                    score += 8

            if row["Wicket"] == 1:
                if temp["wickets"].any() == 2:
                    score += 4
                elif temp["wickets"].any() == 3:
                    score += 4
                elif temp["wickets"].any() == 5:
                    score += 8

            wic = row["Wicket"] + temp["wickets"]
            runs = row["Bowler Runs Conceded"] + temp["runs_conceded"]
            balls = row["Valid Ball"] + temp["balls_bowled"]

            new_row = {
                "match_id": match_id,
                "date": row["Date"],
                "bowler": bowler,
                "wickets": wic,
                "wicket_score": score,
                "balls_bowled": balls,
                "runs_conceded": runs,
                "maiden": maiden,
            }

            df_bowl.loc[condition, new_row.keys()] = new_row.values()

        else:
            if pd.notna(row["Method"]) and row["Method"] != "run out":
                score = 25
                if (row["Method"] == "lbw") | (row["Method"] == "bowled"):
                    score += 8
            else:
                score = 0

            new_row = {
                "match_id": match_id,
                "date": row["Date"],
                "bowler": bowler,
                "wickets": row["Wicket"],
                "wicket_score": score,
                "balls_bowled": row["Valid Ball"],
                "runs_conceded": row["Bowler Runs Conceded"],
                "maiden": maiden,
            }
            if df_bowl.empty:
                df_bowl = pd.DataFrame([new_row], columns=bowling_columns.keys())
            else:
                df_bowl = pd.concat([df_bowl, pd.DataFrame([new_row])], ignore_index=True)
                df_bowl.head()

    return df_bowl


df_bowl = create_bowling_sheet(df, df_bowl)
df_bowl
```

```python
df_bowl.to_csv("bowling_custom2.csv", index=False)
```

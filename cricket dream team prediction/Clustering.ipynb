## Batting


### Clustering for Opening Batsmen

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler



players_list = [Match1_CSK[0], Match1_CSK[1], Match1_GT[0], Match1_GT[1]]
data = pd.DataFrame()
for player in players_list:
    player_df = pd.DataFrame(df[df['Batter'].str.contains(player, case=False, regex=True)])
    data = pd.concat([data, player_df], ignore_index=True)


batsmen_dict = {row['Batter']: [row['Strike Rate'], row['Batting Average'], row['No. of Innings'], row['Balls Faced per Wicket']] for _, row in data.iterrows()}



selected_batsmen = []

weight_average = 1.7
weight_strike_rate = 2.0
weight_innings = 0.5
weight_balls_per_wicket = 0.6


normalized_data = StandardScaler().fit_transform(data[['Batting Average', 'Strike Rate', 'No. of Innings','Balls Faced per Wicket']])

data['Weighted_Score'] = (
    (weight_average * normalized_data[:, 0]) +
    (weight_strike_rate * normalized_data[:, 1]) +
    (weight_innings * normalized_data[:, 2])+
    (weight_innings * normalized_data[:, 3])
)


num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
data['Cluster'] = kmeans.fit_predict(data[['Weighted_Score']])


cluster_means = data.groupby('Cluster')['Weighted_Score'].mean()
top_clusters = cluster_means.nlargest(2).index


for cluster in top_clusters:
    cluster_batsmen = data[data['Cluster'] == cluster].nlargest(2, 'Weighted_Score')
    selected_batsmen.append(cluster_batsmen)
    if sum(len(batsmen) for batsmen in selected_batsmen) >= 2:
        break


selected_batsmen_df = pd.concat(selected_batsmen).nlargest(2, 'Weighted_Score')
selected_batsmen_names = selected_batsmen_df['Batter'].tolist()


filtered_batsmen_dict_opener = {name: batsmen_dict[name] for name in selected_batsmen_names}
print("Selected Batsmen Dictionary:")
print(filtered_batsmen_dict_opener.keys())
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
for cluster in range(num_clusters):
    cluster_data = data[data['Cluster'] == cluster]
    plt.scatter(cluster_data['Weighted_Score'], [0] * len(cluster_data), label=f'Cluster {cluster + 1}')

plt.title('KMeans Clustering of Batsmen')
plt.xlabel('Weighted Score')
plt.legend()
plt.show()
```

### Clustering For Bowler

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler



exclude_players = []
include_players = [Match1_CSK[8], Match1_CSK[9], Match1_CSK[10] , Match1_CSK[7], Match1_GT[8], Match1_GT[9], Match1_GT[10], Match1_GT[7]]
bowlers_data = dfb[['Bowler', 'Bowling Average', 'Strike Rate', 'Economy Rate', 'Matches Bowled In', 'Total Wickets']]

bowlers_data = bowlers_data[~bowlers_data['Bowler'].isin(exclude_players)]

bowler_dfs = [pd.DataFrame(bowlers_data[bowlers_data['Bowler'].str.contains(player, case=False, regex=True)]) for player in include_players]
data = pd.concat(bowler_dfs, ignore_index=True)
bowlers_dict = {row['Bowler']: [row['Bowling Average'], row['Strike Rate'], row['Economy Rate'], row['Matches Bowled In'], row['Total Wickets']] for _, row in data.iterrows()}


selected_bowlers = []

# Define weights for features
weight_average = 0.4
weight_strike_rate = 0.4
weight_economy_rate = 0.2

weight_matches_bowled = 0.3
weight_total_wickets = 0.4


normalized_data = StandardScaler().fit_transform(data[['Bowling Average', 'Strike Rate', 'Economy Rate', 'Matches Bowled In', 'Total Wickets']])

# Calculate Weighted Score
data['Weighted_Score'] = (
    (weight_average * normalized_data[:, 0]) +
    (weight_strike_rate * normalized_data[:, 1]) +
    (weight_economy_rate * normalized_data[:, 2]) +
    (weight_matches_bowled * normalized_data[:, 3]) +
    (weight_total_wickets * normalized_data[:, 4])
)


num_clusters = 4


kmeans = KMeans(n_clusters=num_clusters, random_state=42)
data['Cluster'] = kmeans.fit_predict(data[['Weighted_Score']])

# Identify the Cluster with the Overall Lowest Average Weighted Score
overall_cluster_means = data.groupby('Cluster').mean()
selected_clusters = overall_cluster_means.nsmallest(num_clusters, 'Weighted_Score').index

# Select Bowlers from the Lowest Clusters
for cluster in selected_clusters:
    cluster_bowlers = data[data['Cluster'] == cluster].nlargest(1, 'Weighted_Score')
    selected_bowlers.append(cluster_bowlers)

selected_bowlers_data = pd.concat(selected_bowlers)

print("Selected Bowlers:")
print(selected_bowlers_data[['Bowler', 'Bowling Average', 'Strike Rate', 'Economy Rate', 'Matches Bowled In', 'Total Wickets']])

bowler = list(selected_bowlers_data['Bowler'])
print(bowler)
```
